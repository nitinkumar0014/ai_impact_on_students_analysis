```python
## AI Usage Category
```

```python
def categorize_ai_usage(hours):
    if hours <= 5:
        return 'Low'
    elif hours <= 15:
        return 'Moderate'
    else:
        return 'Heavy'

df['AI_Usage_Category'] = df['Weekly_GenAI_Hours'].apply(categorize_ai_usage)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=avg_ai_usage,
    x='Major_Category',
    y='Weekly_GenAI_Hours'
)

plt.title('Average AI Usage by Major')

plt.xlabel('Major Category')
plt.ylabel('Average Weekly AI Hours')

plt.xticks(rotation=45)

plt.show()
```

```python
major_usage = df.groupby(
    'Major_Category'
)['Weekly_GenAI_Hours'].sum()

percentage_usage = (
    major_usage / major_usage.sum()
) * 100

percentage_usage = percentage_usage.reset_index()

percentage_usage.columns = [
    'Major_Category',
    'AI_Usage_Percentage'
]

percentage_usage
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    percentage_usage['AI_Usage_Percentage'],
    labels=percentage_usage['Major_Category'],
    autopct='%1.1f%%'
)

plt.title('AI Usage Percentage by Major')

plt.show()
```

```python
from scipy.stats import ttest_ind
low_users = df[df['AI_Usage_Category'] == 'Low']['Post_Semester_GPA']

heavy_users = df[df['AI_Usage_Category'] == 'Heavy']['Post_Semester_GPA']

#T-test
t_stat, p_value = ttest_ind(low_users, heavy_users)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Significant difference found")
else:
    print("No significant difference")
```

```python
#Now check WHICH group has higher GPA.
```

```python
print("Low Users GPA Mean:", low_users.mean())

print("Heavy Users GPA Mean:", heavy_users.mean())
```

```python
# An independent t-test was conducted to compare post-semester GPA
# between low AI users and heavy AI users.

# The results showed a statistically significant difference in GPA
# between the two groups (t = 3.49, p < 0.05).

# Low AI users demonstrated a slightly higher average GPA (3.34)
# compared to heavy AI users (3.32).

# This suggests that excessive AI usage may not necessarily improve
# academic performance and could potentially reduce learning
# effectiveness when overused.
```
